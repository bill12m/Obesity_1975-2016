```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette('Paired')
```

```python
#The following functions are for breaking up larger regions into subregions and graphing the data of each one.

def assign_regions(df, list_of_lists):
    for index in df.index:
        for region in list_of_lists:
            for country in region:
                if df.at[index,'Country'] == country:
                    df.at[index,'Region'] = region[-1]
    return(df)
    
def graph_regions(df,region):
    g = sns.relplot(data = df, kind = 'line',
               x = 'Year', y = 'BMI', hue = 'Region', ci = None)
    g.fig.suptitle('Change in BMI for '+region)
    #I want this one to have a different palette for each column, but haven't figured it out yet.
    #sns.relplot(data = df, kind = 'line',
    #           x = 'Year', y = 'BMI', hue = 'Country', col = 'Region', ci = None)
    
    h = sns.catplot(data = df.mask(df['Sex'] == 'Both sexes').dropna(), kind = 'swarm', 
                    x = 'Sex', y = 'BMI',
                    hue = 'Year', ci = None, legend = None, palette = 'Blues')
    h.fig.suptitle('Distribution of BMI by Gender for '+region)
    sns.catplot(data = df.mask(df['Sex'] == 'Both sexes').dropna(), kind = 'swarm',
                    x = 'Sex', y = 'BMI',
                    hue = 'Year', col = 'Region', ci = None, legend = None, palette = 'Reds')

#The following function is for casting string attributes to numerics
def casting_strings(df):
    countries = dict(zip(df['Country'].unique(),range(len(df['Country'].unique()))))
    regions = dict(zip(df['Region'].unique(),range(len(df['Region'].unique()))))
    genders = dict(zip(df['Sex'].unique(), range(len(df['Sex'].unique()))))
    
    df['Country ID'] = df['Country'].apply(lambda x: countries.get(x) if x in countries else x)
    df['Region ID'] = df['Region'].apply(lambda x: regions.get(x) if x in regions else x)
    df['Sex ID'] = df['Sex'].apply(lambda x: genders.get(x) if x in genders else x)
    
#The following function is for adding a decade column to a dataframe
def set_decade(df):
    for index in df.index:
        if df.at[index,'Year'] < 1980:
            df.at[index,'Decade'] = '1970s'
        elif df.at[index,'Year'] < 1990:
            df.at[index,'Decade'] = '1980s'
        elif df.at[index,'Year'] < 2000:
            df.at[index,'Decade'] = '1990s'
        elif df.at[index,'Year'] < 2010:
            df.at[index,'Decade'] = '2000s'
        else:
            df.at[index,'Decade'] = '2010s'
    return df
```

```python
data_clean = pd.read_csv('data_clean.csv')

#data_clean['Continent'] = data_clean['Region']
asia = ['South Asia', 'Middle East', 'Southeast Asia', 'East Asia', 'Central Asia']
europe = ['Balkans', 'Eastern Europe', 'Southern Europe', 'Western Europe', 'Northern Europe']
africa = ['North Africa', 'Southern Africa', 'West Africa', 'East Africa', 'Central Africa']
north_america = ['Caribbean', 'Central America', 'North America']
south_america = ['South America']
oceania = ['Oceania']

#del index
multi_index = pd.MultiIndex.from_frame(data_clean[['Continent','Country','Region']])
data_clean = data_clean.set_index(multi_index)

data_clean = data_clean.drop(columns = ['Continent', 'Region'])
print(data_clean.head())
```

# Asia

The following graphs were my first attempts to gather information about the data. I originally graphed line and swarm plots for every continent and region, but only kept the set for Asia as a display for what these plots look like.

```python
for region in asia:
    df_region = data_clean.xs(['Asia', str(region)], level = [0,2])
    g = sns.relplot(data = df_region, kind = 'line',
               x = 'Year', y = 'BMI', hue = 'Country', ci = None)
    g.fig.suptitle('Change in BMI for '+str(region))
    
    df_region = df_region.mask(df_region['Sex'] == 'Both sexes').dropna()
    h = sns.catplot(data = df_region, kind = 'swarm',
                   x = 'Sex', y = 'BMI', hue = 'Year', palette = 'Blues', legend = None, order = ['Male', 'Female'])
    h.fig.suptitle('Distribution of BMI by Gender for  '+str(region))
```

# Oceania

The graphs below help visualize the obesity data for Oceania, which is a region of the world known for it's high rates of obesity. The line plot breaks it up into subregions and tracks the change in BMI for each one. The first swarm plot compare the BMI's of each gender across the entire region, and is of the same type as what was shown in the Asia section. The smaller swarm plots show how each subregion contributes to the total result. The points of the swarm plots get darker according to the year.

```python
oceania = data_clean.xs('Oceania', level = 2).drop('Country', axis = 1).reset_index()
melanesia = ['Fiji', 'Papua New Guinea', 'Solomon Islands', 'Vanuatu','Melanesia']
micronesia = ['Micronesia (Federated States of)', 'Kiribati', 'Nauru', 'Palau', 'Micronesia']
polynesia = ['Cook Islands', 'Marshall Islands', 'Niue', 'Samoa', 'Tonga', 'Tuvalu', 'Polynesia']
big_islands = ['Australia', 'New Zealand', 'Big Islands']
list_of_lists = [melanesia, micronesia, polynesia, big_islands]

oceania = assign_regions(oceania, list_of_lists)
graph_regions(oceania,'Oceania')
del oceania
```

# Europe

Below is the first scatterplot matrix I built for this project. The grid itself if a pre-made one from seaborn. The `covariance_matrix` function mostly just casts the string types to integers so they'll appear in the matrix.

This is the first time I can see that BMI is approximately normally distributed. which isn't a surprise, but something I haven't looked into yet on the other continents. Also, there is something going on between BMI and the Country/Region which I hadn't seen before. I don't know what it is; but if I can isolate the year, I might find something.

Otherwise, everything is as I expected it to be, there's not much to be found in the covariances when everything but BMI is categorical. I only wanted to check if I was missing any patterns between BMI and something else.

```python
europe = data_clean.xs('Europe', level = 0).drop('Country', axis = 1).reset_index()
europe = europe.mask(europe['Country'] == 'Monaco').mask(europe['Country'] == 'San Marino').mask(europe['Sex'] == 'Both sexes').dropna()
casting_strings(europe)
g = sns.pairplot(data = europe)
```

```python
europe = set_decade(europe)

g = sns.FacetGrid(data = europe, col = 'Decade')
g.map(sns.pointplot, 'Region', 'BMI', order = europe['Region'].unique())
g.fig.suptitle('BMI in each Region of Europe by Decade', y = 1.05)

for ax in g.axes.flat:
    labels = ax.get_xticklabels() # get x labels
    ax.set_xticklabels(labels, rotation=90) # set new labels
plt.show()
del europe
```

## Looking into the relationship between BMI and location over time

So, above we can see that Southern Europe has always had a larger BMI in relation to the other European regions, but it seems like the growth in BMI has been constant across each region of Europe. Something much more interesting is what is shown below, where the Middle East starts off having the largest BMI of any Asian region, and then increases its BMI faster than any other Asian region.

```python
asia = data_clean.xs('Asia', level = 0).drop('Country', axis = 1).reset_index()
asia = set_decade(asia)

g = sns.FacetGrid(data = asia, col = 'Decade')
g.map(sns.pointplot, 'Region', 'BMI', order = asia['Region'].unique())
g.fig.suptitle('BMI in each Region of Asia by Decade', y = 1.05)

for ax in g.axes.flat:
    labels = ax.get_xticklabels() # get x labels
    ax.set_xticklabels(labels, rotation=90) # set new labels
plt.show()
del asia
```
